```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('/content/customer_data.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create histograms for 'age' and 'income'
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Age histogram
sns.histplot(df['age'], bins=20, kde=True, ax=axes[0], color='blue')
axes[0].set_title('Age Distribution')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Frequency')

# Income histogram
sns.histplot(df['income'], bins=20, kde=True, ax=axes[1], color='green')
axes[1].set_title('Income Distribution')
axes[1].set_xlabel('Income')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python


# Analyze the relationship between income and purchase amount using scatter plots
plt.figure(figsize=(10, 6))
sns.scatterplot(x='income', y='purchase_amount', data=df)
plt.title('Relationship between Income and Purchase Amount')
plt.xlabel('Income')
plt.ylabel('Purchase Amount')
plt.show()

# Investigate how different factors like loyalty status and region influence the satisfaction score
plt.figure(figsize=(12, 6))
sns.boxplot(x='loyalty_status', y='satisfaction_score', data=df)
plt.title('Satisfaction Score by Loyalty Status')
plt.xlabel('Loyalty Status')
plt.ylabel('Satisfaction Score')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='region', y='satisfaction_score', data=df)
plt.title('Satisfaction Score by Region')
plt.xlabel('Region')
plt.ylabel('Satisfaction Score')
plt.show()

# Identify and plot a box plot to show any outliers in purchase_amount
plt.figure(figsize=(10, 6))
sns.boxplot(y='purchase_amount', data=df)
plt.title('Box Plot of Purchase Amount')
plt.ylabel('Purchase Amount')
plt.show()
```
