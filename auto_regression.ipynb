```python
# Load the macrodata dataset and store it in a DataFrame
df = sm.datasets.macrodata.load_pandas().data

# Calculate the 'month' column based on the 'quarter' column
df['month'] = (df["quarter"] - 1) * 3 + 1

# Select the desired columns, including the newly calculated 'month' column
df = df[['realgdp', 'realcons', 'realinv', 'month']]

df.head()
```

```python
df['data'] = pd.to_datetime(df[['year','month']].assign(DAY=1))
df.set_index('data',inplace=True)
df.drop(['year','month'],axis=1,inplace=True)
df.head()
```

```python
y=df['realgdp']
```

```python
from statsmodels.tsa.ar_model import AutoReg
model=AutoReg(y,lags=12)
```

```python
model_fit=model.fit()
```

```python
prediction= model_fit.predict(start=len(y),end=len(y)+10)
print(prediction)
```

```python
plt.plot(y,label='realgdp')
plt.plot(prediction,label='prediction')
plt.legend()
plt.show()
```
