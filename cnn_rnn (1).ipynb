````python
def create_factor_dataframes(factor_names_index_dict:dict, df:pd.DataFrame):
  """

  Args:
    factor_names_index_dict:
    df:
  """

  keys = list(factor_names_index_dict.keys())
  data_frames = []
  for name in factor_names:
    if name in keys[:-1]:
        next_key_index = keys.index(name) + 1
        next_key = keys[next_key_index]
    else:
        next_key = None
    factor_df = df[int(factor_names_index_dict[name][0]):int((factor_names_index_dict[next_key][0]) - 1) if next_key != None else len(df)]
    factor_df.Factors.fillna(factor_df.Factors.mode()[0],inplace=True)
    factor_df.SE.fillna(factor_df.SE.mode()[0],inplace=True)
    factor_df = factor_df.reset_index(drop=True)
    data_frames.append(factor_df)

  return data_frames
````

````python
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load data
data_path = '/content/combined_data.csv'
data = pd.read_csv(data_path)

# Define features and target
features = data.drop('C/S( Cs )', axis=1)
target = data['C/S( Cs )'].apply(lambda x: 1 if x < 1 else 0)  # Binary classification

# Identifying numeric and categorical features
numeric_features = features.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = features.select_dtypes(include=['object']).columns.tolist()

# Preprocessing for numeric and categorical data
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Create the modeling pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', RandomForestClassifier())])

# Splitting data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Model training and evaluation
pipeline.fit(X_train, y_train)
model_accuracy = pipeline.score(X_test, y_test)
model_accuracy
````

````python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, recall_score, precision_score, f1_score

# Prediction on test data
y_pred = pipeline.predict(X_test)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()

# Evaluation metrics
recall = recall_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Recall: {recall:.2f}")
print(f"Precision: {precision:.2f}")
print(f"F1 Score: {f1:.2f}")
````

````python
# Function to calculate and display metrics and confusion matrix
def evaluate_model(model_name, y_test, y_pred):
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()

    # Evaluation metrics
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    print(f"{model_name} Recall: {recall:.2f}")
    print(f"{model_name} Precision: {precision:.2f}")
    print(f"{model_name} F1 Score: {f1:.2f}")


print("RandomForest")
# Evaluate RandomForest model
evaluate_model("RandomForest", y_test, y_pred_rf)


print("XGboost")
# Evaluate XGBoost model
evaluate_model("XGBoost", y_test, y_pred_xgb)
````

````python
# Deep Learning.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_curve, auc, accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, GRU, Conv1D, MaxPooling1D, Flatten, Dropout,BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam, SGD, RMSprop, Adagrad, Adadelta, Nadam

# Define features and target
features = data.drop('C/S( Cs )', axis=1)
target = data['C/S( Cs )'].apply(lambda x: 1 if x < 1 else 0)  # Binary classification

# Identifying numeric and categorical features
numeric_features = features.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = features.select_dtypes(include=['object']).columns.tolist()

# One-hot encoding for categorical features
encoder = OneHotEncoder(handle_unknown='ignore')
encoded_categorical_data = encoder.fit_transform(features[categorical_features]).toarray()
encoded_categorical_features = encoder.get_feature_names_out(categorical_features)

# Scaling numeric features
scaler = StandardScaler()
scaled_numeric_data = scaler.fit_transform(features[numeric_features])

# Combine processed numeric and categorical data
processed_features = np.concatenate([scaled_numeric_data, encoded_categorical_data], axis=1)

# Reshape data for LSTM, RNN, and GRU (samples, timesteps, features)
# Assume each feature is a timestep
X = processed_features.reshape(processed_features.shape[0], processed_features.shape[1], 1)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.3, random_state=42)

# Convert targets to categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, GRU, Conv1D, MaxPooling1D, Flatten, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam, SGD, RMSprop, Adagrad, Adadelta, Nadam

def build_model(model_type='LSTM', units=32, dropout_rate=0.2, optimizer_name='adam', learning_rate=0.001, additional_dense_layers=False, use_batch_norm=False, momentum=0.0, nesterov=False):
    model = Sequential()

    # Setting up the input layer based on model type
    if model_type == 'LSTM':
        model.add(LSTM(units, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False))
        if use_batch_norm:
            model.add(BatchNormalization())
    elif model_type == 'RNN':
        model.add(SimpleRNN(units, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False))
        if use_batch_norm:
            model.add(BatchNormalization())
    elif model_type == 'GRU':
        model.add(GRU(units, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=False))
        if use_batch_norm:
            model.add(BatchNormalization())
    elif model_type == 'CNN':
        model.add(Conv1D(units, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
        model.add(MaxPooling1D(pool_size=2))
        if use_batch_norm:
            model.add(BatchNormalization())
        model.add(Flatten())

    # Conditional dropout layer
    model.add(Dropout(dropout_rate))

    # Conditional additional dense layer
    if additional_dense_layers:
        model.add(Dense(64, activation='relu'))
        model.add(Dropout(dropout_rate))

    # Final Dense layer configuration
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(2, activation='softmax'))  # Output layer for binary classification formatted as multiclass

    # Configuring the optimizer with variable learning rate
    optimizer_dict = {
        'adam': Adam(learning_rate=learning_rate),
        'sgd': SGD(learning_rate=learning_rate),
        'rmsprop': RMSprop(learning_rate=learning_rate),
        'adagrad': Adagrad(learning_rate=learning_rate),
        'adadelta': Adadelta(learning_rate=learning_rate),
        'nadam': Nadam(learning_rate=learning_rate)
    }
    model.compile(optimizer=optimizer_dict[optimizer_name], loss='binary_crossentropy', metrics=['accuracy'])
    return model
````

````python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, roc_curve, auc
from tensorflow.keras.optimizers import Adam, SGD, RMSprop, Adagrad, Adadelta, Nadam

def train_and_evaluate_with_hyperparams(model_types, params_grid, X_train, y_train_cat, X_test, y_test):
    best_accuracy = 0
    best_model_key = ""
    best_model = None
    best_y_pred = None
    best_params = {}

    for model_type in model_types:
        for units in params_grid['units']:
            for dropout_rate in params_grid['dropout_rate']:
                for learning_rate in params_grid['learning_rate']:
                    for optimizer_name in params_grid['optimizer_name']:
                        for momentum in params_grid.get('momentum', [0]):  # Default to [0] if not in params
                            for nesterov in params_grid.get('nesterov', [False]):  # Default to [False] if not in params
                                key = f"{model_type}_{optimizer_name}_units={units}_lr={learning_rate}_mom={momentum}_nest={nesterov}"
                                print(f"Training model: {key}")
                                model = build_model(model_type=model_type, units=units, dropout_rate=dropout_rate,
                                                    optimizer_name=optimizer_name, learning_rate=learning_rate,
                                                    momentum=momentum, nesterov=nesterov)
                                accuracy, y_pred, fpr, tpr, roc_auc = train_and_record(model, X_train, y_train_cat, X_test, y_test)
                                if accuracy > best_accuracy:
                                    best_accuracy = accuracy
                                    best_model_key = key
                                    best_model = model
                                    best_y_pred = y_pred  # Update the best prediction here
                                    best_params = {
                                        'model_type': model_type,
                                        'units': units,
                                        'dropout_rate': dropout_rate,
                                        'learning_rate': learning_rate,
                                        'optimizer_name': optimizer_name,
                                        'momentum': momentum,
                                        'nesterov': nesterov
                                    }

    # Save the best model and plot the confusion matrix if a best model has been found
    if best_model:
        best_model.save(f"best_model_{best_model_key}.h5")

        # Ensure best_y_pred is not None before plotting
        if best_y_pred is not None:
            cm = confusion_matrix(y_test, best_y_pred)
            disp = ConfusionMatrixDisplay(confusion_matrix=cm)
            disp.plot()
            plt.title(f'Best Model Confusion Matrix: {best_model_key}')
            plt.show()

    return {
        'best_accuracy': best_accuracy,
        'best_fpr': fpr,
        'best_tpr': tpr,
        'best_roc_auc': roc_auc,
        'best_params': best_params
    }

def train_and_record(model, X_train, y_train_cat, X_test, y_test):
    # Train the model
    model.fit(X_train, y_train_cat, epochs=10, batch_size=32, validation_split=0.2, verbose=2)

    # Predict probabilities for the test set
    y_pred_prob = model.predict(X_test)

    # Check the shape of y_test to determine if it's already in label form
    if len(y_test.shape) == 2 and y_test.shape[1] > 1:  # Assuming one-hot encoded labels
        y_true = np.argmax(y_test, axis=1)
    else:
        y_true = y_test  # Assuming y_test is already in label form

    # Extract predicted labels from probabilities
    y_pred = np.argmax(y_pred_prob, axis=1)

    # Calculate accuracy
    accuracy = accuracy_score(y_true, y_pred)

    # Calculate ROC curve and AUC, assuming binary classification
    if y_pred_prob.shape[1] > 1:  # Check if the predictions include probabilities for each class
        fpr, tpr, _ = roc_curve(y_true, y_pred_prob[:, 1])  # Probability for the positive class
        roc_auc = auc(fpr, tpr)
    else:  # If only one probability is given per sample, use it directly
        fpr, tpr, _ = roc_curve(y_true, y_pred_prob[:, 0])
        roc_auc = auc(fpr, tpr)

    return accuracy, y_pred, fpr, tpr, roc_auc


````
